## Ejercicio 003 - Listas, Funciones y Easter Eggs
**Capítulo:** 1.9 Primeros pasos con Python 
**Objetivo:** Crear listas, definir funciones, usar `for` y probar easter eggs de Python

Este ejercicio combina 4 conceptos: almacenamiento de datos en listas, 
reutilización de código con funciones, iteración con bucles y el humor de Python.

```python
"""
Módulo: 003_listas_funciones.py
Descripción: Practica listas, funciones propias y easter eggs de Python
"""

def calcular_media(lista_numeros):
    """Recibe una lista de números y regresa su promedio aritmético"""
    if len(lista_numeros) == 0:
        return 0 # Evita error de división entre cero
    return sum(lista_numeros) / len(lista_numeros)

def main():
    """Función principal que ejecuta todos los ejemplos del ejercicio"""
    
    # 1. Listas: almacenamos edades
    edades = [18, 25, 32, 62] 
    print(f"Lista de edades: {edades}")
    print(f"Media directa: {sum(edades) / len(edades)}")
    
    # 2. Funciones: usamos nuestra función propia
    media_resultado = calcular_media(edades)
    print(f"Media con función: {media_resultado}")
    
    # 3. Bucles for: iteramos la lista
    print("\nRecorriendo edades con for:")
    for edad in edades:
        print(f"Tiene {edad} años")
    
    # 4. Peculiaridad de Python: división
    print(f"\nPython divide bien: 2 / 4 = {2/4}")
    
    # 5. Easter eggs: descomenta para probar
    # import this
    # import antigravity

if __name__ == "__main__":
    main()
```
