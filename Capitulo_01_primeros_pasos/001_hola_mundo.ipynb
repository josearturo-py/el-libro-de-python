Este es mi primer programa en Python. Demuestra el uso básico de salida por pantalla.

```python
"""
Módulo: 01_hola_mundo.py
Descripción: Imprime el mensaje clásico "Hola Mundo" y una variante personalizada
"""

def mostrar_saludo():
    """Imprime saludos básicos usando print()"""
    # 1. Saludo básico del libro
    mensaje_basico = "Hola mundo"
    print(mensaje_basico)
    
    # 2. Extra: Usando f-strings para personalizar
    nombre = "Arturo"
    ciudad = "Colima"
    print(f"¡Hola Mundo desde {ciudad}! Soy {nombre} y estoy aprendiendo Python")

if __name__ == "__main__":
    mostrar_saludo()
```
