## Ejercicio 002 - Variables str e int
**Capítulo:** 1.9 Primeros pasos con Python  
**Libro:** *El Libro de Python* - Álvaro Revuelta  
**Objetivo:** Declarar variables de tipo `str` e `int` y mostrarlas con f-strings  

En este ejercicio practico la creación de variables de texto y números enteros.
Uso f-strings porque son la forma más moderna y legible desde Python 3.6+.

```python
"""
Módulo: 002_variables_str_int.py
Descripción: Ejemplo de variables string e integer usando f-strings
"""

def practicar_variables():
    """Demuestra el uso de variables str e int con datos personales"""
    
    # Ejemplo 1: Datos del libro
    nombre_libro = "Alicia"
    edad_libro = 20
    print(f"Ejemplo del libro: {nombre_libro} tiene {edad_libro} años")
    
    print("-" * 30)  # Línea separadora
    
    # Ejemplo 2: Mis datos personales
    nombre_usuario = "Arturo"
    edad_usuario = 62  
    ciudad_usuario = "Colima"
    estado_civil = "Soltero"  # Extra: agregué otra str
    
    # f-string con varias variables
    print(f"Me llamo {nombre_usuario}, tengo {edad_usuario} años")
    print(f"Soy de {ciudad_usuario} y actualmente estoy {estado_civil}")
    
    # Bonus: Verificamos tipos de dato
    print(f"\nVerificación de tipos:")
    print(f"La variable 'nombre_usuario' es: {type(nombre_usuario)}")
    print(f"La variable 'edad_usuario' es: {type(edad_usuario)}")

if __name__ == "__main__":
    practicar_variables()
```
