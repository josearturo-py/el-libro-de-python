```python
""" Los bool almacenan un tipo de datos que solo
    toman dos valores: True o False """
    
b = True
print(b)    # True
print(type(b))
```

```python
"""Pyton permite convertir a bool cualquier cosa interpretando:
    True : Cuando no es cero o no está vacío
    False: Cuando es cero o está vacío """
    
print (bool(10))    # True
print (bool(0))     # False
print (bool([]))    # False
print (bool([1]))   # True
print (bool(None))  # False
```

```python
# Gracias a que da falso cuanto está vacío se puede hacer lo siguiente:

lista = []
if not lista:
    print ("Lista vacía")
```

```python
# Curiosidad: Python interpreta True como 1

print (True + True)  # 2
print (True + 10)    # 11
```

```python
# Como usar los operadores lógicos

print (True and True)   # True
print (True and False)  # False
print (False and False) # False
print (True or True)    # True
print (True or False)   # True
print (False or False)  # False
print (not True)        # False
print (not False)       # True
```

```python
# Soporta los operadores matemáticos

a, b = True, True

print (a,b)    # True True

print (a+b)    # 2
print (a-b)    # 0
print (a*b)    # 1
print (a/b)    # 1-0
print (a**b)   # 1
```
