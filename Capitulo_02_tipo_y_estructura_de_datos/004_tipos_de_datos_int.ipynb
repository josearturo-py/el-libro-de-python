## Tipos de Datos: Enteros

**Objetivo:** Entender cómo Python maneja números enteros, memoria dinámica y diferentes bases numéricas.

**Conceptos clave:**
1. Los `int` no tienen límite fijo como en C/Java
2. Python asigna memoria según el tamaño del número
3. Se pueden declarar en decimal, binario, octal y hexadecimal

```python
# Los int permiten almacear números sin contenido decimal

i = 12
print (i)     
# 12

# Con type podemos ver el tipo de variable

print (type(i))   
# <class 'int'>
```

```python
""" Vamos a importar sys para ver el tamaño del entero en bytes y como ocupa
    mas espacio conforme crece. """

import sys
print (f"El número 1 ocupa {(sys.getsizeof (1))} bytes")
print (f"El número 10 ocupa {(sys.getsizeof (10))} bytes")
print (f"El número 100 ocupa {(sys.getsizeof (100))} bytes")
print (f"El número 1000 ocupa {(sys.getsizeof (1000))} bytes")
print (f"El número 10000 ocupa {(sys.getsizeof (10000))} bytes")
print (f"El número 10000 ocupa {(sys.getsizeof (100000))} bytes")
print (f"El número {(250**250)} ocupa {(sys.getsizeof (250**250))} bytes")
print (f"El número 5**10000 ocupa {(sys.getsizeof (5**10000))} bytes")
```

# Veamos ahora usando un ciclo for como un entero en Phyton ocupa más memoria conforme crece:

for i in [1, 100, 10000, 10**100, 10**1000]:
    print (f"{i} ocupa {sys.getsizeof(i)} bytes")

```python
# Los int soportan las operaciones comunes, recordemos que ** es exponente

a, b = 3, 6
print (a+b)   # 9
print (a-b)   # -3
print (a*b)   # 18
print (a/b)   # 0.5
print (a**b)  # 729
```

### Bases Numéricas en Python

Un `int` se puede declarar en diferentes bases. Python lo convierte automáticamente a decimal internamente, pero todas son `<class 'int'>`.

- **Binario:** prefijo `0b` → `0b11001` = 25
- **Octal:** prefijo `0o` → `0o31` = 25  
- **Hexadecimal:** prefijo `0x` → `0x19` = 25
- **Decimal:** sin prefijo → `25` = 25

```python
# Los int pueden ser declarados tanto en decimal, binario, hexadecial y octal

a = 0b11001 # 25 en binario
b = 0x19    # 25 en hexadecimal
c = 0o031   # 25 en octal
d = 25      # 25 en decimal

print (a, type(a)) # 25 <class 'int'>
print (b, type(b)) # 25 <class 'int'>
print (c, type(c)) # 25 <class 'int'>
print (d, type(d)) # 25 <class 'int'>
```

### Conclusiones del ejercicio

1. **Memoria dinámica:** `sys.getsizeof()` demuestra que un `int` crece en bytes conforme el número es más grande.
2. **Sin overflow:** `250**250` o `5**10000` no truena como en otros lenguajes.
3. **Múltiples bases:** `0b`, `0o`, `0x` son solo formas de escribir el mismo `int`.
4. **Operadores:** `**` es exponente, `/` siempre da `float`, usa `//` para división entera.
