Este notebook demuestra que los enteros en Python no tienen un límite fijo de 32 o 64 bits como en otros lenguajes. Usamos `sys.getsizeof()` para medir cómo crece el objeto `int` en memoria y probamos literales `0b`, `0o`, `0x` para distintas bases numéricas.

```python
def main():
    # Los int permiten almacenar números sin contenido decimal
    i = 12
    print(i)  # 12
    print("-"*60)
    
    # Con type podemos ver el tipo de variable
    print(type(i))  # <class 'int'>
    
    '''
    En Python existe solo una clase de int y permite almacenar valores muy grandes
    Se asigna memoria según sea necesario.
    '''
    x = 250**250
    print(x)
    print("-"*60)
    
    """ 
    Vamos a importar sys para ver el tamaño del entero en bytes 
    y como ocupa más espacio conforme crece.
    """
    import sys
    print(f"El número 1 ocupa {sys.getsizeof(1)} bytes")
    print(f"El número 10 ocupa {sys.getsizeof(10)} bytes")
    print(f"El número 100 ocupa {sys.getsizeof(100)} bytes")
    print(f"El número 1000 ocupa {sys.getsizeof(1000)} bytes")
    print(f"El número 10000 ocupa {sys.getsizeof(10000)} bytes")
    print(f"El número 100000 ocupa {sys.getsizeof(100000)} bytes")
    print(f"El número {250**250} ocupa {sys.getsizeof(250**250)} bytes")
    print(f"El número 5**10000 ocupa {sys.getsizeof(5**10000)} bytes")
    print("-"*60)
    
    # Veamos ahora usando un ciclo for como un entero en Python ocupa más memoria conforme crece:
    for i in [1, 100, 10000, 10**100, 10**1000]:
        print(f"{i} ocupa {sys.getsizeof(i)} bytes")
    print("-"*60)
    # Los int soportan las operaciones comunes, recordemos que ** es exponente
    a, b = 3, 6
    print(f"a+b = {(a+b)}")  # 9
    print(f"a-b = {(a-b)}")  # -3
    print(f"a*b = {(a*b)}")  # 18
    print(f"a/b = {(a/b)}")  # 0.5
    print(f"a**b = {(a**b)}")  # 729
    print("-"*60)
    # Los int pueden ser declarados tanto en decimal, binario, hexadecimal y octal
    a = 0b11001  # 25 en binario
    b = 0x19     # 25 en hexadecimal
    c = 0o031    # 25 en octal
    d = 25       # 25 en decimal 
    print(f"a={a}, b={b}, c={c}, d={d}")
    
if __name__ == "__main__":
    main()
```
