```python
## Capítulo 2.1.2 - float
**Libro:** *El Libro de Python* - Álvaro Revuelta  
**Objetivo:** El tema busca que se entienda que los float son útiles para decimales, pero no son matemáticamente exactos. Saber esto evita bugs cuando se comparen números o se hagan cálculos financieros.  
**Fecha:** 2026-05-26  

Este notebook demuestra que los float en Python sirven para almacenar números decimales, pero tienen limitaciones importantes que debes conocer.
```

```python
def main():
    ## Tema: Números Float en Python
    
    # Los números float permiten almacenar números decimales
    f = 0.35
    print("Los números float permiten almacenar números decimales:\n", f)
    print(type(f))  # <class 'float'>
    print("-" * 60)
    
    # También admiten notación científica
    f = 1.93e-3
    print("También admiten notación científica (1.93e-3):\n", f)
    print("-" * 60)
    
    # La precisión no es infinita
    f = 0.99999999999999999
    print("La precisión no es infinita: 1 == ", f)
    print(1 == f)  # True
    print("-" * 60)
    
    # También puede almacenar el valor infinito
    x = float('inf')
    print(x)
    print(type(x))  # <class 'float'>
    print("-" * 60)
    
    # Como la precisión no es infinita, hay números que no se pueden almacenar exactamente
    f = 1.3
    print(f)
    print(f"{f:.20}")  # 1.3000000000000000444
    print("-" * 60)
    
    # Es posible redondear un float de varias maneras
    import math
    f = 2.675
    print("Redondeo de 2.675:")
    print("round(f, 2) ->", round(f, 2))  # 2.67
    print("math.ceil(f) ->", math.ceil(f))  # 3
    print("math.floor(f) ->", math.floor(f))  # 2
    print("-" * 60)
    
    # Los float tienen operaciones comunes
    a, b = 3.3, 6.5
    print("Operaciones con a=3.3 y b=6.5:")
    print("a + b =", a + b)  # 9.8
    print("a - b =", a - b)  # -3.2
    print("a * b =", a * b)  # 21.45
    print("a / b =", a / b)  # 0.5076923076923077
    print("a ** b =", a ** b)  # 2346.0680721890503

if __name__ == "__main__":
    main()
```
