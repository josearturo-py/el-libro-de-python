```python
# Los números float permiten almacenar números decimales

f = 0.35
print (f)        # 0.35
print (type(f))  # <class 'float'>
```

```python
# También admiten notación científica

f = 1.93e-3
```

```python
# La precisión no es infinita
# En este caso f no es realmente 1.0, pero se almacena como 1
# (al decimal 17 se redondea)
f = 0.99999999999999999
print (f)        # 1.0
print (1 == f)   # true
```

```python
# También puede almacenar el valor infinito

x = float ('inf')
print (x)           # inf
print (type(x))     # <class 'float'>
```

```python
# Como la precisión no es infinita, hay números que 
# no se pueden almacenar exactamente, por ejemplo 1.3

f = 1.3
print (f)           # 1.3
print (f"{f:.20}")  # 1.3000000000000000444
```

```python
# Es posible redondear un float de varias maneras

import math

f = 2.675
print (round(f, 2))     # 2.67
print (math.ceil(f))    # 3
print (math.floor(f))   # 2
```

```python
# los float tienen opreraciones comunes

a, b = 3.3, 6.5

print(a+b)  # 9.8
print(a-b)  # -3.2
print(a*b)  # 21.45
print(a/b)  # 0.5076923076923077
print(a**b) # 2346.0680721890503
```
