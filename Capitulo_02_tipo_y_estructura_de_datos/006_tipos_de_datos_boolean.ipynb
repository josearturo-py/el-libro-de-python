**Objetivo:** Entender que `bool` es el tipo lГіgico de Python con solo dos valores: `True` y `False`. Base para condicionales `if`, bucles y operadores lГіgicos. 
**Fecha:** 2026-05-26

Los `bool` permiten almacenar un tipo que solo toma dos valores: `True` o `False`.

```python
def main():
    """ Los bool almacenan un tipo de datos que solo toman dos valores: True o False """
    b = True
    print(f"Valor de b: {b}") # Valor de b: True
    print(f"Tipo de b: {type(b)}") # Tipo de b: <class 'bool'>

    """Python permite convertir a bool cualquier cosa interpretando:
    True : Cuando no es cero o no estГЎ vacГ­o
    False: Cuando es cero o estГЎ vacГ­o """
    print(f"bool(10) в†’ {bool(10)}") # bool(10) в†’ True
    print(f"bool(0) в†’ {bool(0)}") # bool(0) в†’ False
    print(f"bool([]) в†’ {bool([])}") # bool([]) в†’ False
    print(f"bool([1]) в†’ {bool([1])}") # bool([1]) в†’ True
    print(f"bool(None) в†’ {bool(None)}") # bool(None) в†’ False

    # Gracias a que da falso cuando estГЎ vacГ­o se puede hacer lo siguiente:
    lista = []
    if not lista:
        print("La lista estГЎ vacГ­a porque bool([]) es False")

    # Como usar los operadores lГіgicos
    print(f"True and True в†’ {True and True}") # True and True в†’ True
    print(f"True and False в†’ {True and False}") # True and False в†’ False
    print(f"False and False в†’ {False and False}") # False and False в†’ False
    print(f"True or True в†’ {True or True}") # True or True в†’ True
    print(f"True or False в†’ {True or False}") # True or False в†’ True
    print(f"False or False в†’ {False or False}") # False or False в†’ False
    print(f"not True в†’ {not True}") # not True в†’ False
    print(f"not False в†’ {not False}") # not False в†’ True

    # Curiosidad: Python interpreta True como 1
    print(f"True + True в†’ {True + True}") # True + True в†’ 2
    print(f"True + 10 в†’ {True + 10}") # True + 10 в†’ 11

    # Soporta los operadores matemГЎticos
    a, b = True, True
    print(f"a={a}, b={b}") # a=True, b=True
    print(f"a+b в†’ {a+b}") # a+b в†’ 2
    print(f"a-b в†’ {a-b}") # a-b в†’ 0
    print(f"a*b в†’ {a*b}") # a*b в†’ 1
    print(f"a/b в†’ {a/b}") # a/b в†’ 1.0
    print(f"a**b в†’ {a**b}") # a**b в†’ 1

if __name__ == "__main__":
    main()
```
