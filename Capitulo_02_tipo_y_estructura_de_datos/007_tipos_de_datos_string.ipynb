Los `str` (cadenas) permiten almacenar texto. Pueden escribirse con comillas dobles, comillas simples o tres comillas.## Capítulo 2.1.4 - string

```python
def main():
    
    """ Los string permiten almacenar texto se pueden escribir con comillas, comillas simples o tres comillas """
    s = "El libro de Python"
    print(f"Valor de s: {s}") # Valor de s: El libro de Python
    print(f"Tipo de s: {type(s)}") # Tipo de s: <class 'str'>
    
    s = 'El libro de Python'
    print(f"Usando comillas simples: {s}") # Usando comillas simples: El libro de Python
    s = """El libro
    de 
    Python"""
    print(f"Usando tres comillas: {s}") # Usando tres comillas: El libro de Python
    
    # Hay múltiples formas de combinar str con otros tipos de datos
    # con int
    x = 5
    s = "El número es: %d" % x
    print(f"Usando %d: {s}") # Usando %d: El número es: 5
    
    # También con % como en C
    s = "Los números son %d y %d." % (5, 10)
    print(f"Usando %d múltiples: {s}") # Usando %d múltiples: Los números son 5 y 10.
    
    # Con format
    s = "Los números son {} y {}.".format(5, 10)
    print(f"Usando format: {s}") # Usando format: Los números son 5 y 10.
    
    # Con format y el nombre de la variable
    s = "Los números son {a} y {b}.".format(a=5, b=10)
    print(f"Usando format con nombres: {s}") # Usando format con nombres: Los números son 5 y 10.
    
    # LA PREFERIDA CON fstring introducida en la PEP 498
    a, b = 5, 10
    print(f"Usando f-string: Los números son {a} y {b}") # Usando f-string: Los números son 5 y 10

if __name__ == "__main__":
    main()
```
