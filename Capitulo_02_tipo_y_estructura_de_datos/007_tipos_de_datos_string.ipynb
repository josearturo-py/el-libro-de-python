```python
""" Los string permiten almacenar texto
    se pueden escribir con comillas, comillas simples o tres comillas """

s = "El libro de Python"
print (s)                   # El libro de Python
print (type(s))             # <class 'str'"

s = 'El libro de Python'
print ("""El libro
de
Python""")
```

```python
# Hay múltiples formas de combiar str con otos tipos de datos

# con int
x = 5
s = "El número es: %d" % x
print (s)                    # El número es: 5

# También con % como en C
s = "Los números son %d y %d." % (5, 10)
print (s)   # Los números son 5 y 10.

# Con format
s = "Los números son {} y {}.".format(5, 10)
print (s)   # Los números son 5 y 10.

# Con format y el nombre de la variable
s = "Los números son {a} y {b}.".format(a=5, b=10)
print (s)   # Los números son 5 y 10.

# LA PREFERIDA CON fstring introducida en la PEP 498
a, b = 5, 10
s = f"Los números son {a} y {b}"
print (s)
```
